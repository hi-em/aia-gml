## 7. Primal Graph

```python
cluster = Cluster.ByTopologies(cc)
vertices = Topology.Vertices(cluster)
print(vertices)
vertices = [Topology.Copy(v) for v in vertices]
edges = Topology.Edges(cluster)
edges = [Topology.Copy(e) for e in edges]
primal_graph = Graph.ByVerticesEdges(vertices , edges)
```

```python
vertices = Graph.Vertices(primal_graph)
for v in vertices:
    d = Dictionary.ByKeysValues(["size", "color"], [18, "red"])
    v = Topology.SetDictionary(v, d)
edges = Graph.Edges(primal_graph)
for e in edges:
    d = Dictionary.ByKeysValues(["width", "color"], [4, "black"])
    e = Topology.SetDictionary(e, d)
```

```python
Topology.Show(cc, primal_graph,
              vertexSizeKey="size",
              vertexColorKey="color",
              edgeWidthKey="width",
              edgeColorKey="color",
              faceOpacity=0.3,
              backgroundColor="white",
              width=500,
              height=500,
              renderer=renderer)
```

## 8. Dual Graph

```python
dual_graph = Graph.ByTopology(cc, direct=True)
```

```python
vertices = Graph.Vertices(dual_graph)
for v in vertices:
    d = Dictionary.ByKeysValues(["size", "color"], [18, "red"])
    v = Topology.SetDictionary(v, d)
edges = Graph.Edges(dual_graph)
for e in edges:
    d = Dictionary.ByKeysValues(["width", "color"], [4, "black"])
    e = Topology.SetDictionary(e, d)
```

```python
Topology.Show(cc, dual_graph,
              vertexSizeKey="size",
              vertexColorKey="color",
              edgeWidthKey="width",
              edgeColorKey="color",
              faceOpacity=0.3,
              backgroundColor="white",
              width=500,
              height=500,
              renderer=renderer)
```

## 9. Importing the doors

```python
door_objs = Topology.ByOBJPath(r"C:\Users\chidi\OneDrive\Desktop\macad\module-03\aia-gml\assets\doors.obj", 
    selfMerge=True)
door_topos = door_objs if isinstance(door_objs, list) else [door_objs]
doors = Cluster.Faces(Cluster.ByTopologies(door_topos))
print("Door faces:", len(doors))
```

```python
cc_with_doors = Topology.AddApertures(
    cc,
    doors,
    exclusive=True,
    subTopologyType="face"
 )
dual_graph = Graph.ByTopology(cc_with_doors, direct=False, viaSharedApertures=True)
print("Dual graph edges:", len(Graph.Edges(dual_graph)))
```

```python
vertices = Graph.Vertices(dual_graph)
for v in vertices:
    d = Dictionary.ByKeysValues(["size", "color"], [18, "red"])
    v = Topology.SetDictionary(v, d)
edges = Graph.Edges(dual_graph)
for e in edges:
    d = Dictionary.ByKeysValues(["width", "color"], [4, "black"])
    e = Topology.SetDictionary(e, d)
```

```python
Topology.Show(cc_with_doors, doors, dual_graph,
              vertexSizeKey="size",
              vertexColorKey="color",
              edgeWidthKey="width",
              edgeColorKey="color",
              faceOpacity=0.3,
              backgroundColor="white",
              width=500,
              height=500,
              renderer=renderer)
```
